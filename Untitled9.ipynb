```python

import pandas as pd
import yfinance as yf
import requests
from io import StringIO
import matplotlib.pyplot as plt
```

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data["Date"] = tesla_data["Date"].dt.tz_localize(None)
tesla_data.head()
```

```python
url_tesla = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
response_tesla = requests.get(url_tesla, headers=headers)
html_io_tesla = StringIO(response_tesla.text)
tables_tesla = pd.read_html(html_io_tesla)

tesla_revenue = tables_tesla[1]
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue.dropna(inplace=True)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "", regex=False).str.replace(",", "", regex=False)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(float)
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])
tesla_revenue.head()
```

```python
df_tesla = pd.merge_asof(
    tesla_data.sort_values("Date"),
    tesla_revenue.sort_values("Date"),
    on="Date"
)
df_tesla.tail()
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

ax1.set_xlabel("Date")
ax1.set_ylabel("Close Price (USD)", color="blue")
ax1.plot(df_tesla["Date"], df_tesla["Close"], color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
ax2.set_ylabel("Revenue (Millions USD)", color="green")
ax2.plot(df_tesla["Date"], df_tesla["Revenue"], color="green")
ax2.tick_params(axis="y", labelcolor="green")

plt.title("Tesla: Stock Price vs Revenue")
plt.show()
```

```python
# Dados histricos da GameStop
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data["Date"] = gme_data["Date"].dt.tz_localize(None)

# Dados de receita da GameStop (via scraping)
url_gme = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
response_gme = requests.get(url_gme, headers=headers)
html_io_gme = StringIO(response_gme.text)
tables_gme = pd.read_html(html_io_gme)

gamestop_revenue = tables_gme[1]
gamestop_revenue.columns = ["Date", "Revenue"]
gamestop_revenue.dropna(inplace=True)
gamestop_revenue["Revenue"] = gamestop_revenue["Revenue"].str.replace("$", "", regex=False).str.replace(",", "", regex=False)
gamestop_revenue["Revenue"] = gamestop_revenue["Revenue"].astype(float)
gamestop_revenue["Date"] = pd.to_datetime(gamestop_revenue["Date"])

# Merge dos dados
df_gme = pd.merge_asof(
    gme_data.sort_values("Date"),
    gamestop_revenue.sort_values("Date"),
    on="Date"
)
df_gme.tail()
```

```python
 fig, ax1 = plt.subplots(figsize=(12,6))

ax1.set_xlabel("Date")
ax1.set_ylabel("Close Price (USD)", color="red")
ax1.plot(df_gme["Date"], df_gme["Close"], color="red")
ax1.tick_params(axis="y", labelcolor="red")

ax2 = ax1.twinx()
ax2.set_ylabel("Revenue (Millions USD)", color="orange")
ax2.plot(df_gme["Date"], df_gme["Revenue"], color="orange")
ax2.tick_params(axis="y", labelcolor="orange")

plt.title("GameStop: Stock Price vs Revenue")
plt.show()
```
